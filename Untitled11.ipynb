```python
# 1) Find usable JPGs in your Archive (skip __MACOSX and ._ files)
all_imgs = glob.glob("/content/drive/MyDrive/Archive/**/*.jpg", recursive=True)
real_imgs = [
    p for p in all_imgs
    if "__MACOSX" not in p and "/._" not in p and os.path.getsize(p) > 5000
]

print("Total jpg found:", len(all_imgs))
print("Real usable jpg:", len(real_imgs))
assert len(real_imgs) >= 60, "Not enough real images found in Archive."

# 2) Generate 60 Rochester coordinates
LAT_MIN, LAT_MAX = 43.10, 43.23
LON_MIN, LON_MAX = -77.72, -77.52

def generate_rochester_points(n=60, seed=42):
    random.seed(seed)
    return pd.DataFrame(
        [(random.uniform(LAT_MIN, LAT_MAX), random.uniform(LON_MIN, LON_MAX)) for _ in range(n)],
        columns=["lat", "lon"]
    )

coords_df = generate_rochester_points(60)

# 3) Choose 60 real images + combine with coords
selected_imgs = real_imgs[:60]
img_df = pd.DataFrame({"image_path": selected_imgs})

combined_df = pd.concat([img_df, coords_df], axis=1)
print("combined_df size:", len(combined_df))
combined_df.head()
```

```python
# Use fallback pretrained weights you already have
base_w = "/content/yolov8s-seg.pt"
model = YOLO(base_w)
print("Model loaded:", base_w)

DEC_PRED_DIR = "/content/dec_geo"
os.makedirs(DEC_PRED_DIR, exist_ok=True)

model.predict(
    source=combined_df["image_path"].tolist(),
    save=True,
    save_txt=True,
    conf=0.30,
    project=DEC_PRED_DIR,
    name="pred",
    exist_ok=True
)

PRED_LABELS_DIR = f"{DEC_PRED_DIR}/pred/labels"
print("Labels saved at:", PRED_LABELS_DIR)
print("Num label files:", len(glob.glob(PRED_LABELS_DIR + "/*.txt")))

# --- weights per image ---
def parse_conf(line):
    parts = line.strip().split()
    try:
        return float(parts[-1])
    except:
        return 0.0

def image_weight(img_path):
    base = os.path.splitext(os.path.basename(img_path))[0]
    lab = os.path.join(PRED_LABELS_DIR, base + ".txt")
    if not os.path.exists(lab):
        return 0.0
    confs = []
    with open(lab, "r") as f:
        for line in f:
            confs.append(parse_conf(line))
    return float(np.mean(confs)) if confs else 0.0

combined_df["weight"] = combined_df["image_path"].apply(image_weight)

# This is what your sweep needs:
dec_geo_df = combined_df[combined_df["weight"] > 0].copy()
print("‚úÖ dec_geo_df size (detections only):", len(dec_geo_df))
dec_geo_df.head()
```

```python
# ----------------------------------------
# FUNCTION: Build Heatmap + Clusters
# ----------------------------------------
def build_weighted_heatmap_clusters(df, out_html, eps=0.003, min_samples=8):
    center = [df["lat"].mean(), df["lon"].mean()]
    m = folium.Map(location=center, zoom_start=12)

    heat_data = df[["lat", "lon", "weight"]].values.tolist()
    HeatMap(
        heat_data,
        radius=18, blur=25, min_opacity=0.5,
        gradient={0.0:"#0000ff", 0.4:"#00ff00", 0.7:"#ffa500", 1.0:"#ff0000"},
    ).add_to(m)

    coords = df[["lat", "lon"]].values
    db = DBSCAN(eps=eps, min_samples=min_samples).fit(coords)
    df["cluster"] = db.labels_

    for c in sorted(df["cluster"].unique()):
        if c == -1:
            continue
        sub = df[df["cluster"] == c]
        folium.CircleMarker(
            location=[sub["lat"].mean(), sub["lon"].mean()],
            radius=9,
            color="#18a558",
            fill=True,
            fill_opacity=0.85,
            popup=f"Cluster {c} (n={len(sub)})",
        ).add_to(m)

    cmap = cm.LinearColormap(["blue", "green", "orange", "red"], vmin=0, vmax=1)
    cmap.caption = "Confidence-weighted trash density"
    m.add_child(cmap)
    m.add_child(LatLngPopup())

    m.save(out_html)
    print(f"‚úÖ Heatmap saved: {out_html}")
    return df


# ----------------------------------------
# STEP 1 ‚Äî Run sweep
# ----------------------------------------
eps_values = [0.003, 0.004, 0.005, 0.006, 0.008, 0.010]
min_samples_values = [3, 4, 5, 6, 8]

results = []
print("üîç Running DBSCAN sweep...\n")

coords = dec_geo_df[["lat", "lon"]].values

for eps in eps_values:
    for ms in min_samples_values:
        labels = DBSCAN(eps=eps, min_samples=ms).fit(coords).labels_
        clusters, counts = np.unique(labels, return_counts=True)
        cluster_info = dict(zip(clusters, counts))
        results.append((eps, ms, cluster_info))
        print(f"eps={eps:0.3f}, min_samples={ms} ‚Üí clusters: {cluster_info}")

# ----------------------------------------
# STEP 2 ‚Äî pick best config (first that forms any real cluster)
# ----------------------------------------
chosen_eps, chosen_ms = None, None
for eps, ms, cluster_info in results:
    if any(k >= 0 for k in cluster_info.keys()):
        chosen_eps, chosen_ms = eps, ms
        break

if chosen_eps is None:
    chosen_eps, chosen_ms = 0.010, 3   # fallback

print("\nüéØ BEST PARAMETERS SELECTED:")
print("eps =", chosen_eps)
print("min_samples =", chosen_ms)

# ----------------------------------------
# STEP 3 ‚Äî build final map
# ----------------------------------------
OUT_MAP = "/content/drive/MyDrive/Archive/runs/litter_heatmap_clusters_december_second.html"

geo_df_out = build_weighted_heatmap_clusters(
    dec_geo_df.copy(),
    OUT_MAP,
    eps=chosen_eps,
    min_samples=chosen_ms,
)

print("\nüìå Final Cluster Counts:")
print(geo_df_out["cluster"].value_counts())
```
